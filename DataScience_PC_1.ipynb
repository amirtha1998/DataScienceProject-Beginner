```python
#this function segregates the highly correlated columns with respect to threshold value
#dataset is X data where the target variable is removed
#threshold is the value that we choose
#col_corr is the column names which are highly correlated with respect to the threshold

def  correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
      for j in range(i):
        if abs(corr_matrix.iloc[i,j]) > threshold:
          colname = corr_matrix.columns[i]
          col_corr.add(colname)
    return col_corr
```
